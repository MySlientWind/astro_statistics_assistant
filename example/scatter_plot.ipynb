```python
import numpy as np
import matplotlib.pyplot as plt
from asa.dataset import Dataset
```

```python
x = np.random.normal(size=1000)
y = x**2 + np.random.normal(size=1000)
z = np.log(np.abs(x + y)) + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z]).T
```

```python
dataset = Dataset(data, ['x', 'y', 'z'], ['x', 'y', 'z'])
```

```python
dataset.scatter(
    'x',
    ['y', 'z'],
    #range=([[-3, 3], [-10, 10]]),
    subplots_kwargs={'figshape': (1, 2)},
    plot_kwargs={
        's': 1,
    })
```

```python
fig, axes = plt.subplots(1,3,figsize=(16,5))
dataset.scatter(
    'x',
    'y',
    plot_kwargs={
        's': 10,
    },
    axes=axes[0])

dataset.scatter(
    'x',
    'y',
    plot_kwargs={
        's': 10,
    },
    z_name='z',
    ifsmooth = False,
    label='test',
    axes=axes[1])

dataset.scatter(
    'x',
    'y',
    plot_kwargs={
        's': 100,
        'edgecolors' : 'black',
    },
    axes=axes[2])

dataset.scatter(
    'x',
    'y',
    plot_kwargs={
        's': 50,
        'edgecolors' : 'none',
    },
    z_name='z',
    ifsmooth = True,
    smooth_kwargs={
        'nsmooth' : 0.3,
    },
    label='test',
    axes=axes[2])
```
