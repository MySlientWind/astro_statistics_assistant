```python
import numpy as np
from asa.Bcorner import corner
from asa.dataset import Dataset
```

```python
x = np.random.normal(size=1000)
y = x**2 + np.random.normal(size=1000)
z = np.log(np.abs(x + y)) + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z, x > 0, np.abs(y)]).T
dataset = Dataset(data, ['x', 'y', 'z', 'xg0', 'absy'], ['x_label', 'y_label', 'z_label', 'xg0_label', 'absy_label'])
```

```python
dataset.corner(
    ['x', 'y', 'z'],
    kde_smooth=True,
    kde_smooth1d=True
)
```

```python
dataset.corner(
    ['x', 'y', 'z'],
    smooth=1,
    smooth1d=1
)
```

```python
dataset.corner(['x', 'y', 'z'], kde_smooth=True, labels=None)
```

```python
dataset.corner(kde_smooth=True)
```

```python
corner(np.array([x, y, z]).T, kde_smooth=True);
```
