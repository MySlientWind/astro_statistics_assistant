```python
import numpy as np
import matplotlib.pyplot as plt
from asa.dataset import Dataset
```

```python
x = np.random.normal(size=1000)
y = x**2 + np.random.normal(size=1000)
z = np.log(np.abs(x + y)) + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z, x > 0, np.abs(y), y**2]).T
```

```python
dataset = Dataset(data, ['x', 'y', 'z', 'xg0', 'absy', 'y2'], ['x_l', 'y_l', 'z_l', 'xg0_l', 'absy_l', 'y2_l'])
```
