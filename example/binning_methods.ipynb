```python
import numpy as np
import matplotlib.pyplot as plt

from asa.weighted_statistic import mean as wmean
from asa.weighted_statistic import get_effect_sample_size
from asa.binning_methods import get_stat_method, bin_1d
```

```python
x = np.random.normal(size=100)
y = 3 * x + np.random.normal(size=100)
w = np.random.uniform(size=100)
```

```python
center, edges, bin_index, statistic = bin_1d(x,
                                                y,
                                                weights=w,
                                                x_statistic=[
                                                    'mean', 'median',
                                                    'std', 'std_mean',
                                                    'std_median', 'q:0.3',
                                                    'q:0.7'
                                                ],
                                                y_statistic=[
                                                    'mean', 'median',
                                                    'std', 'std_mean',
                                                    'std_median', 'q:0.3',
                                                    'q:0.7'
                                                ],
                                                range=(-3, 3))
```

```python
plt.scatter(x, y, c='gray', alpha=0.5)
plt.errorbar(statistic['x_mean'], statistic['y_mean'], xerr=statistic['x_std_mean'], yerr=statistic['y_std_mean'])
plt.errorbar(center, statistic['y_mean'], xerr=(center[1] - center[0])/2, yerr=statistic['y_std_mean'])
```
