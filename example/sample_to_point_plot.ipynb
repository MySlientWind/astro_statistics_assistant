```python
import numpy as np
import matplotlib.pyplot as plt
from asa.dataset import Dataset
from asa.Bcorner import quantile
```

```python
x = np.random.normal(size=1000)
y = x**2 + np.random.normal(size=1000)
z = np.log(np.abs(x + y)) + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z, x > 0, np.abs(y), y**2]).T
```

```python
dataset = Dataset(data, ['x', 'y', 'z', 'xg0', 'absy', 'y2'], ['x_label', 'y_label', 'z_label', 'xg0_label', 'absy_label', 'y2_label'])
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

dataset.contour(['x'] * 6, 'y', axes=axes, plot_density=False)

dataset.sample_to_point(
    ['x'] * 6,
    'y',
    center_type_each=['mean', 'mean', 'mean', 'median', 'median', 'median'],
    error_type_each=[
        'std', 'std_mean', 'quantile', 'std', 'std_mean', 'quantile'
    ],
    axes=axes)
```
