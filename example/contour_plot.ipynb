```python
import numpy as np
from asa.dataset import Dataset
```

```python
x = np.random.normal(size=1000)
y = x**2 + np.random.normal(size=1000)
z = np.log(np.abs(x + y)) + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z, x > 0]).T
```

```python
dataset = Dataset(data, ['x', 'y', 'z', 'xg0'], ['x', 'y', 'z', 'xg0'])
```

```python
dataset.contour('x', ['y', 'z'], subplots_kwargs={'figshape': (1, 2)})
```

```python
dataset.contour('x', ['y', 'z'], subplots_kwargs={'figshape': (1, 2)}, color_each=['red', 'blue'])
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample=x > 0)
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample='xg0')
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample='x>0')
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample='-1.5<x<1.5')
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample=x > 0, kde_smooth=True)
```

```python
dataset.contour('x', ['y', 'z'],
                subplots_kwargs={'figshape': (1, 2)},
                color_each=['red', 'blue'],
                subsample=x > 0, smooth=2)
```
