```python
import numpy as np
import matplotlib.pyplot as plt
from asa.dataset import Dataset
```

```python
x = np.random.normal(size=1000)
y = np.random.normal(size=1000)
z = x**2 + y**2 + np.random.normal(size=1000)
```

```python
data = np.array([x, y, z]).T
dataset = Dataset(data, ['x', 'y', 'z'], ['x_label', 'y_label', 'z_label'])
```

```python
dataset.heatmap('x',
                'y',
                bins=11,
                range=[[-3, 3], [-3, 3]],
                z_name='z',
                pcolor_kwargs={'cmap': 'jet'})
```

```python
dataset.heatmap('x',
                'y',
                bins=11,
                range=[[-3, 3], [-3, 3]],
                z_name='z',
                map_kind='contour',
                title='',
                contour_kwargs={'cmap': 'jet'})
```

```python
dataset.heatmap('x',
                'y',
                bins=11,
                range=[[-3, 3], [-3, 3]],
                z_name='z',
                map_kind='contour',
                contour_kwargs={'cmap': 'jet'},
                set_clabel=True)
```
